```python
plt.figure(figsize=(8,5))
sns.histplot(df['category'], bins=30, kde=True)
plt.title("Distribution of Sports Footwear Sales")
plt.xlabel("Count")
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['day'] = df['order_date'].dt.day
```

```python
df['revenue_usd'].sum()
```

```python
sns.scatterplot(
    x='discount_percent',
    y='final_price_usd',
    data=df
)
plt.title("Discount vs Final Price")
plt.show()
```

```python
sns.barplot(
    x='category',
    y='revenue_usd',
    data=df,
    estimator=np.sum
)
plt.xticks(rotation=45)
plt.title("Revenue by Category")
plt.show()
```

```python
country_revenue = df.groupby('country')['revenue_usd'].sum().sort_values(ascending=False)

country_revenue.head(10).plot(kind='bar')
plt.title("Top Countries by Revenue")
plt.show()
```

```python
sns.countplot(x='payment_method', data=df)
plt.title("Payment Method Distribution")
plt.show()
```

```python
sns.boxplot(
    x='customer_income_level',
    y='final_price_usd',
    data=df
)
plt.title("Income Level vs Spending")
plt.show()
```

```python
sns.boxplot(y=df['revenue_usd'])
plt.title("Revenue Outliers")
plt.show()
```

```python
sns.histplot(df['revenue_usd'], bins=30, kde=True)
plt.title("Revenue Distribution")
plt.show()
```
